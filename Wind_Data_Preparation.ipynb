# Baron


```python
ds = xr.open_dataset('/content/drive/MyDrive/TA (1)/Processing Data/Data/raw data/baron_raw.nc')  # sesuaikan disini
ds
```

```python
df = ds.to_dataframe()
df
```

```python
df2=df.reset_index()
df2
```

```python
#filtering based on lat lon
new_df = df2.loc[(df2['longitude']==110.5) & (df2['latitude']==-8.00) ]
new_df
```

```python
new_df.dtypes
```

```python

#filtering based on date
new_df = new_df.loc[(df2['time'] >= '2014-07-14 15:00:00')
                     & (df2['time'] < '2014-09-05 04:00:00')]
```

```python
new_df
```

```python
#calculate magnitude
import math
new_df['mag'] = new_df.apply(lambda x: math.sqrt(x['u10']**2 +x['v10']**2),axis=1)
new_df
```

```python
new_df = new_df[['time','mag']]
new_df
```

```python
new_df.to_csv('/content/drive/MyDrive/TA (1)/Processing Data/Data/raw data/baron_angin.csv')
```
